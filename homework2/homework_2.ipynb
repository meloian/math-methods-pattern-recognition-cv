### Colour Balancing
In this notebook we will show different type of colour balancing making use of von Kries' hypothesis.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [15, 5]
```

```python
img = cv2.imread('sea.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

### White patch
In white patch algorithm we choose a group of pixels we know they should be white. We then scale the resulting image colour channels by this white patch.

```python
# Define white patch and the coefficients
row, col = 485, 864
white = img[row, col, :]
coeffs = 255.0/white

# Apply white balancing and generate balanced image
balanced = np.zeros_like(img, dtype=np.float32)
for channel in range(3):
    balanced[..., channel] = img[..., channel] * coeffs[channel]

# White patching does not guarantee that the dynamic range is preserved, images must be clipped.
balanced = balanced/255
balanced[balanced > 1] = 1

plt.subplot(121), plt.imshow(img)
plt.subplot(122), plt.imshow(balanced)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
balanced_gray = cv2.cvtColor((balanced*255).astype(np.uint8), cv2.COLOR_RGB2GRAY)

# Compute histograms
hist_orig, _ = np.histogram(img_gray.ravel(), bins=256, range=(0,255))
hist_bal, _ = np.histogram(balanced_gray.ravel(), bins=256, range=(0,255))

# Plot histograms
plt.subplot(121), plt.plot(hist_orig), plt.title("Original Histogram"), plt.grid(True)
plt.subplot(122), plt.plot(hist_bal), plt.title("Balanced Histogram"), plt.grid(True)
plt.show()
```

### Gray world
This algorithm assumes that a scene, on average, is gray.

```python
# Load your image
img = cv2.imread('dark.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Compute the mean values for all three colour channels (red, green, blue)
mean_r = np.mean(img[..., 0])
mean_g = np.mean(img[..., 1])
mean_b = np.mean(img[..., 2])

# Compute the coefficients kr, kg, kb
# Note: there are 3 coefficients to compute but we only have 2 equations.
# Therefore, you have to make an assumption, fix the value of one of the
# coefficients and compute the remining two
# Hint: You can fix the coefficient of the brightest colour channel to 1.
mean_vals = np.array([mean_r, mean_g, mean_b])
mean_gray = np.mean(mean_vals)

kr = mean_gray / mean_r
kg = mean_gray / mean_g
kb = mean_gray / mean_b

# Apply color balancing and generate the balanced image
balanced = np.zeros_like(img, dtype=np.float32)
balanced[..., 0] = img[..., 0] * kr
balanced[..., 1] = img[..., 1] * kg
balanced[..., 2] = img[..., 2] * kb

balanced = balanced / 255
balanced[balanced > 1] = 1

# Show the original and the balanced image side by side
plt.subplot(121), plt.imshow(img)
plt.subplot(122), plt.imshow(balanced)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
balanced_gray = cv2.cvtColor((balanced*255).astype(np.uint8), cv2.COLOR_RGB2GRAY)

# Compute histograms
hist_orig, _ = np.histogram(img_gray.ravel(), bins=256, range=(0,255))
hist_bal, _ = np.histogram(balanced_gray.ravel(), bins=256, range=(0,255))

# Plot histograms
plt.subplot(121), plt.plot(hist_orig), plt.title("Original Histogram"), plt.grid(True)
plt.subplot(122), plt.plot(hist_bal), plt.title("Balanced Histogram"), plt.grid(True)
plt.show()
```

### Scale-by-max
This is a straightforward algorithm that scales each colour channel by its maximum value. Note that it is sensitive to noise and saturations.

```python
# Load your image
img = cv2.imread('kodim05.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Compute the maximum values for all three colour channels (red, green, blue)
max_r = np.max(img[..., 0])
max_g = np.max(img[..., 1])
max_b = np.max(img[..., 2])

# Apply scale-by-max balancing and generate the balanced image
balanced = np.zeros_like(img, dtype=np.float32)
balanced[..., 0] = img[..., 0] * (255.0 / max_r)
balanced[..., 1] = img[..., 1] * (255.0 / max_g)
balanced[..., 2] = img[..., 2] * (255.0 / max_b)

balanced = balanced / 255
balanced[balanced > 1] = 1

# Show original and balanced images
plt.subplot(121), plt.imshow(img),
plt.subplot(122), plt.imshow(balanced),
plt.show()
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
balanced_gray = cv2.cvtColor((balanced*255).astype(np.uint8), cv2.COLOR_RGB2GRAY)

# Compute histograms
hist_orig, _ = np.histogram(img_gray.ravel(), bins=256, range=(0,255))
hist_bal, _ = np.histogram(balanced_gray.ravel(), bins=256, range=(0,255))

# Plot histograms
plt.subplot(121), plt.plot(hist_orig), plt.title("Original Histogram"), plt.grid(True)
plt.subplot(122), plt.plot(hist_bal), plt.title("Balanced Histogram"), plt.grid(True)
plt.show()
```
